```python
from ultralytics import YOLO


data = {
    'train': 'dataset/images/train/',
    'val': 'dataset/images/val/',
    'annotations': {
        'train': 'dataset/annotations/instances_train.json',
        'val': 'dataset/annotations/instances_val.json'
    }
}


model = YOLO('yolov8x-seg.pt')

model.train(
    data=data,
    epochs=50,
    imgsz=640,
    batch=16,
    workers=8,
    device=0
)


model.save('yolov8-custom.pt')
```

```python
#import cv2
import os

def video_to_frames(video_path, output_folder):
    # Kreiranje izlaznog foldera ako ne postoji
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Otvaranje video fajla
    vidcap = cv2.VideoCapture(video_path)
    success, image = vidcap.read()
    count = 0

    while success:
        # Sačuvaj trenutni frejm kao JPEG sliku
        #Football batch 3.mp4_00000.jpg
        cv2.imwrite(os.path.join(output_folder, f"New video.mp4_{count:05d}.jpg"), image)
        success, image = vidcap.read()
        count += 1

    vidcap.release()
    print(f"Ukupno sačuvano frejmova: {count}")

# Primer korišćenja
#video_path = './archive/Batch-1/Soccer/video/Batch 1.mp4'
#output_folder = './archive/Batch-3/import-videos/images'
video_path = './archive/train-test/test.mp4' # Iseci test
output_folder = './archive/images_test/' #sacuvaj frejmove ovde

video_to_frames(video_path, output_folder)
```
